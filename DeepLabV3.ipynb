```python
import os
import torch
import numpy as np
import torchvision.transforms as T
from torchvision.models.segmentation import deeplabv3_resnet50
from torchvision.models.segmentation import DeepLabV3_ResNet50_Weights
from torch.utils.data import Dataset, DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score

# Configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
#input_size = (256, 256)  # Resize all images
batch_size = 8
num_epochs = 20
num_classes = 5  # Background, Liver, Kidney, Spleen, Pancreas

# Paths to dataset
image_dir = '/kaggle/input/dataset/dataset/train_imgs'
label_dir = "/kaggle/input/dataset/dataset/train_lbs"


# Custom Dataset Class
class SegmentationDataset(Dataset):
    def __init__(self, image_dir, label_dir, transform=None):
        self.image_paths = sorted([os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.png')])
        self.label_paths = sorted([os.path.join(label_dir, f) for f in os.listdir(label_dir) if f.endswith('.npy')])
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load image
        image = cv2.imread(self.image_paths[idx], cv2.IMREAD_GRAYSCALE)  # Grayscale
        #image = cv2.resize(image, input_size)  # Resize
        image = np.repeat(image[:, :, np.newaxis], 3, axis=-1)  # Convert to 3 channels

        # Load label
        label = np.load(self.label_paths[idx])  # Load npy
        #label = cv2.resize(label, input_size, interpolation=cv2.INTER_NEAREST)  # Resize label

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)


# Data Transforms
transform = T.Compose([
    T.ToTensor(),  # Convert to PyTorch Tensor
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Normalize
    #T.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

# Initialize Dataset and Split
dataset = SegmentationDataset(image_dir, label_dir, transform=transform)
num_validation = 100
num_training = len(dataset) - num_validation
train_dataset, val_dataset = random_split(dataset, [num_training, num_validation])

# Data Loaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Load Pretrained DeepLabV3 Model
model = deeplabv3_resnet50(weights=DeepLabV3_ResNet50_Weights.DEFAULT)  # Load with pretrained weights
model.classifier[4] = nn.Conv2d(256, num_classes, kernel_size=1)  # Update the classifier
model.to(device)

# Define Loss and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)


# Training Loop
def train_one_epoch(model, dataloader, optimizer, criterion):
    model.train()
    running_loss = 0.0
    for images, labels in tqdm(dataloader, desc="Training"):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)['out']  # Get model outputs
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    return running_loss / len(dataloader)

# Function to calculate metrics for each class
def calculate_metrics_per_class(outputs, labels, num_classes):
    _, preds = torch.max(outputs, dim=1)  # Get predicted classes
    preds = preds.view(-1).cpu().numpy()  # Flatten predictions
    labels = labels.view(-1).cpu().numpy()  # Flatten ground truth
    
    # Initialize lists to store metrics for each class
    precisions = []
    recalls = []
    f1_scores = []
    
    for class_idx in range(1, num_classes):  # Exclude background class (index 0)
        class_precision = precision_score(labels, preds, labels=[class_idx], average='macro', zero_division=0)
        class_recall = recall_score(labels, preds, labels=[class_idx], average='macro', zero_division=0)
        class_f1 = f1_score(labels, preds, labels=[class_idx], average='macro', zero_division=0)
        
        precisions.append(class_precision)
        recalls.append(class_recall)
        f1_scores.append(class_f1)
    
    return precisions, recalls, f1_scores

# Validation Loop
def validate_one_epoch(model, dataloader, criterion):
    model.eval()
    running_loss = 0.0
    with torch.no_grad():
        for images, labels in tqdm(dataloader, desc="Validation"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)['out']
            loss = criterion(outputs, labels)
            running_loss += loss.item()
    return running_loss / len(dataloader)


# Function to calculate accuracy
def calculate_accuracy(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    correct = (preds == labels).sum().item()
    total = labels.numel()
    return correct / total

# Lists to store training and validation metrics
train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

# Lists to store metrics for each epoch
precision_per_epoch = []
recall_per_epoch = []
f1_per_epoch = []

# Training and Validation
for epoch in range(num_epochs):
    print(f"Epoch {epoch+1}/{num_epochs}")
    
    # Training
    train_loss = train_one_epoch(model, train_loader, optimizer, criterion)
    train_losses.append(train_loss)
    
    # Validation
    val_loss = validate_one_epoch(model, val_loader, criterion)
    val_losses.append(val_loss)
    
    # Compute accuracies
    model.eval()  # Set model to evaluation mode
    train_accuracy = 0
    val_accuracy = 0
    epoch_precisions = []
    epoch_recalls = []
    epoch_f1s = []
    with torch.no_grad():
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)['out']
            train_accuracy += calculate_accuracy(outputs, labels)
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)['out']
            val_accuracy += calculate_accuracy(outputs, labels)
            # Calculate metrics for this batch
            batch_precisions, batch_recalls, batch_f1s = calculate_metrics_per_class(outputs, labels, num_classes)
            epoch_precisions.append(batch_precisions)
            epoch_recalls.append(batch_recalls)
            epoch_f1s.append(batch_f1s)
        # Average metrics over all batches
        epoch_precisions = np.mean(epoch_precisions, axis=0)
        epoch_recalls = np.mean(epoch_recalls, axis=0)
        epoch_f1s = np.mean(epoch_f1s, axis=0)
            
        precision_per_epoch.append(epoch_precisions)
        recall_per_epoch.append(epoch_recalls)
        f1_per_epoch.append(epoch_f1s)
    
    # Average accuracy over all batches
    train_accuracy /= len(train_loader)
    val_accuracy /= len(val_loader)
    
    train_accuracies.append(train_accuracy)
    val_accuracies.append(val_accuracy)

    
    
    print(f"Training Loss: {train_loss:.4f}, Validation Loss: {val_loss:.4f}")
    print(f"Training Accuracy: {train_accuracy:.4f}, Validation Accuracy: {val_accuracy:.4f}")
    print(f"Validation Precision (Organs): {epoch_precisions}")
    print(f"Validation Recall (Organs): {epoch_recalls}")
    print(f"Validation F1-Score (Organs): {epoch_f1s}")

# Save the Model
torch.save(model.state_dict(), "deeplabv3_segmentation_model.pth")

print("Training Complete.")

# Plot Loss and Accuracy
epochs = range(1, num_epochs + 1)
plt.figure(figsize=(12, 5))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(epochs, train_losses, label="Training Loss")
plt.plot(epochs, val_losses, label="Validation Loss")
plt.title("Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, train_accuracies, label="Training Accuracy")
plt.plot(epochs, val_accuracies, label="Validation Accuracy")
plt.title("Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.tight_layout()
plt.show()

# Plot Precision, Recall, and F1-Score
classes = ["Liver", "Kidney", "Spleen", "Pancreas"]
for i, class_name in enumerate(classes):
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, num_epochs + 1), [precision[i] for precision in precision_per_epoch], label="Precision")
    plt.plot(range(1, num_epochs + 1), [recall[i] for recall in recall_per_epoch], label="Recall")
    plt.plot(range(1, num_epochs + 1), [f1[i] for f1 in f1_per_epoch], label="F1-Score")
    plt.title(f"Metrics for {class_name}")
    plt.xlabel("Epochs")
    plt.ylabel("Metric Value")
    plt.legend()
    plt.show()
```

```python
def visualize_predictions(model, dataloader, num_samples=3):
    model.eval()
    samples_shown = 0
    with torch.no_grad():
        for images, labels in dataloader:
            images = images.to(device)
            outputs = model(images)['out']
            _, preds = torch.max(outputs, dim=1)
            
            for i in range(images.shape[0]):
                if samples_shown >= num_samples:
                    return
                plt.figure(figsize=(15, 5))
                plt.subplot(1, 3, 1)
                plt.imshow(images[i].cpu().permute(1, 2, 0).numpy())
                plt.title("Input Image")
                plt.subplot(1, 3, 2)
                plt.imshow(labels[i].cpu().numpy())
                plt.title("Ground Truth")
                plt.subplot(1, 3, 3)
                plt.imshow(preds[i].cpu().numpy())
                plt.title("Prediction")
                plt.show()
                samples_shown += 1

visualize_predictions(model, val_loader, 5)
```

```python
import pandas as pd

# Define test dataset directory
test_image_dir = "/kaggle/input/dataset/dataset/test_imgs"
test_output_file = "submission.csv"

# Load Trained Model
model.load_state_dict(torch.load("deeplabv3_segmentation_model.pth"))
model.to(device)
model.eval()

# Define Run-Length Encoding (RLE) function
def rle_encoding(mask):
    """
    Perform Run-Length Encoding for a given binary mask.
    Args:
        mask: Binary mask (2D numpy array)
    Returns:
        rle: Encoded RLE string
    """
    pixels = mask.flatten()
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return " ".join(map(str, runs))


# Define test dataset
class TestDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_paths = sorted([os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.png')])
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        image_name = os.path.basename(image_path).split(".")[0]

        # Load image
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        #image = cv2.resize(image, input_size)
        image = np.repeat(image[:, :, np.newaxis], 3, axis=-1)

        if self.transform:
            image = self.transform(image)

        return image_name, image


# Test DataLoader
test_transform = T.Compose([
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    #T.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
])

test_dataset = TestDataset(test_image_dir, transform=test_transform)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)

# Generate Submission File
submission = []

for image_name, image in tqdm(test_loader, desc="Predicting"):
    image_name = image_name[0]  # Extract image name
    image = image.to(device)

    # Get model prediction
    with torch.no_grad():
        output = model(image)['out']  # Model output
        preds = torch.argmax(output, dim=1).squeeze(0).cpu().numpy()  # Convert to numpy array

    # Generate RLE for each class
    for class_idx in range(num_classes):
        mask = (preds == class_idx).astype(np.uint8)  # Binary mask for the class
        rle = rle_encoding(mask)
        submission.append({
            "Id": f"{image_name}_{class_idx}",
            "Predicted": rle
        })

# Convert to DataFrame and save as CSV
submission_df = pd.DataFrame(submission)
submission_df.to_csv(test_output_file, index=False)
print(f"Submission file saved to {test_output_file}.")
```
